## Importación de librerias.

````python
import pandas as pd
import numpy as np
````

## Carga de datos

````python
# Ruta del Archivo de datos Excel.
ruta_ventas2024 = "../data/ventas2024.xlsx"

# Cargar las ventas de los meses enero - junio.
ventas2024_ene_jun = pd.read_excel(ruta_ventas2024, sheet_name="ventas_ene-jun")

ventas2024_ene_jun
````

### Extraer datos de meses y ventas por separado.

````python
meses_array = np.array(ventas2024_ene_jun["Mes"]) # Meses. Datos(x).
ventas_array = np.array(ventas2024_ene_jun["Ventas"]) # Ventas. Datos(y).

print(f"Array: Meses.\n {meses_array}\n")
print(f"Array: Ventas.\n {ventas_array}\n")
````

## Operaciones para obtener '**Beta 1**' y '**Beta 2**'.

### Promedios de los datos '**x**' y datos '**y**'.

````python
media_meses = meses_array.mean() # Datos (x).
media_ventas = ventas_array.mean() # Datos (y).

print("Promedios.")

print(f"Meses: {media_meses}")
print(f"Ventas: {media_ventas}")
````

### Errores de '**x**' y de '**y**'.

````python
errores_meses = meses_array - media_meses # Datos (x).
errores_ventas = ventas_array - media_ventas # Datos (y).

print("Errores de meses y ventas.")

print(f"Meses: {errores_meses}")
print(f"Ventas: {errores_ventas}")
````

Errores de '**x**' al cuadrado.

````python
errores_meses_cuadrado = errores_meses**2

print(f"Errores de 'x' al cuadrado: {errores_meses_cuadrado}")

# Sumatoria de los errores de 'x' al cuadrado.

sum_errores_meses_cuadrado = sum(errores_meses_cuadrado)

print(f"Suma de errores de 'x' al cuadrado: {sum_errores_meses_cuadrado}")
````

### Productos entre los errores de '**x**' y '**y**'.

````python
productos_errores_x_y = errores_meses * errores_ventas

print(f"Productos entre los errores 'x' y 'y': {errores_meses_cuadrado}")

# Sumatoria de los productos entre errores 'x' y 'y'.

sum_productos_errores_x_y = sum(productos_errores_x_y)

print(f"Sumatoria de los productos entre errores 'x' y 'y': {sum_productos_errores_x_y}")
````

## Calculo de '**Beta 1**' y '**Beta 0**'.

````python
# Beta 1.

beta1 = sum_productos_errores_x_y/sum_errores_meses_cuadrado

print(f"Beta 1: {beta1}")

# Beta 0.

beta0 = media_ventas - beta1 * media_meses

print(f"Beta 0: {beta0}")
````

## Estimar las siguientes ventas del año.

Ventas predicciones de ventas estimadas para los siguientes meses del año, de julio a diciembre.

````python
ventas2024_jul_dic = []

for mes in range(7, 13):
    y = (beta1 * mes) + beta0
    ventas2024_jul_dic.append(y)

predicciones = pd.DataFrame({
    "Mes": range(7,13),
    "Ventas estimadas": ventas2024_jul_dic
})

predicciones
````
